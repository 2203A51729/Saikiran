```python
import matplotlib.pyplot as plt
people = ['kiran','arun','vijay','varun']
age =[25, 30, 35, 40,]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]
plt.scatter(age, height, color='skyblue', label='Height')
plt.plot(age,height)
plt.scatter(age, weight, color='salmon', label='Weight')
plt.plot(age,weight)
plt.show()
```

```python
plt.bar(people, age, color='skyblue', label='Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution in Healthcare Data')
plt.show()
```

```python
plt.hist(age, bins=5, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution in Healthcare Data')
plt.show()
```
